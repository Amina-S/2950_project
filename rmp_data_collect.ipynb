```python
import sys
import requests
import bs4 
import numpy as np
import pandas as pd
import os
import time
import csv

myfile = open("Rate_my_prof.txt") 
with open("rmp.csv", "w", newline = '') as csvfile:
    writer = csv.writer(csvfile)
    columns = [("Professor", "Quality", "Difficulty", "Would Take Again", "Date", "Class taught")]
    for data in columns:
            writer.writerow(data)
    for line in myfile:
        url = line
        url = url.rstrip("\n")
        page = requests.get(url)
        

        soup = bs4.BeautifulSoup(page.text, "html.parser")
        
        dataset = []
        
        ul      = soup.find('ul', {'class': 'RatingsList__RatingsUL-hn9one-1 kHITzZ'})
        print(url)
        if ul is not None:
            for item in ul.findAll('li', recursive = False):

                if item is not None:
                    
                    date = "None"
                    quality = "None"
                    difficulty = "None"
                    would_take_again = "None"
                    course = "None"
                    
                    
                    name_class = soup.find("div", {"class": 'NameTitle__Name-dowf0z-0 cfjPUG'})
                    first_name = name_class.span.text
                    #last_name_first_name = last_name_class.select('div[class*="NameTitle"]')
                    last_name_class = name_class.find("span", {"class": 'NameTitle__LastNameWrapper-dowf0z-2 glXOHH'})
                    #print(last_name_first_name)
                    last_name = last_name_class.text
                    name = first_name + " " + last_name
                    #print(first_name + " " + last_name)
                  
                    q_d_up_class = item.find("div", {"class": 'RatingValues__StyledRatingValues-sc-6dc747-0 bJSTHc'})                    
                    if q_d_up_class is not None:
                        #print(q_d_up_class.text)
                        #q_up_class = q_d_up_class.find("div", {"class": 'RatingValues__RatingContainer-sc-6dc747-1 DObVa'})                                                    
                        quality_and_difficulty = q_d_up_class.select('div[class*="RatingValues__RatingValue-sc"]')
                        q = str(quality_and_difficulty[0])
                        quality = float(q[-9:-6])
                        d = str(quality_and_difficulty[1])
                        difficulty = float(d[-9:-6])
                        #print(quality, difficulty)
                    
                    would_take_again_up = item.find("div", {"class": 'CourseMeta__StyledCourseMeta-x344ms-0 fPJDHT'})                    
                    if would_take_again_up is not None:
                        would_take_again_list = would_take_again_up.select('div[class*="MetaItem__StyledMetaItem-"]')
                        if len(would_take_again_list) >2:
                            would_take_again = would_take_again_list[2].span.text
                        #print(would_take_again)
                    #RatingHeader__StyledHeader-sc-1dlkqw1-0 kiSEdA
                    time_course_up_class = item.find("div", {"class": 'RatingHeader__StyledHeader-sc-1dlkqw1-0 uBHCj'})
                    if time_course_up_class is not None:
                        time_up_list = time_course_up_class.select('div[class*="TimeStamp__StyledTimeStamp-sc-9q2r30-"]')
                        #print(time_up_list)
                        date = time_up_list[0].text
                        print(date)
                        course_up_list = time_course_up_class.select('div[class*="RatingHeader__ClassInfoWrapper-sc-1dlkqw1-1 jxOApy"]')
                        course_up_class = course_up_list[0].find("div",{"class": 'RatingHeader__StyledClass-sc-1dlkqw1-2 gxDIt'})
                        course = course_up_class.text
                        print(course)
                    

                    dataset.append((name, quality, difficulty, would_take_again, date, course))

            for data in dataset:
                writer.writerow(data)

            time.sleep(0.01)


    
myfile.close()


```
