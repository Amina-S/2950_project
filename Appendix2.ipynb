# Web Scraping Appendix

## [piazza.com](https://piazza.com/)
Information gathered using links and raw data in [piazza_links.txt](piazza_links.txt)

```python
import sys
import requests
import bs4 
import numpy as np
import pandas as pd
import os
import time
import csv

myfile = open("piazza_links.txt") 

with open("piazza.csv", "w", newline = '') as csvfile:
    writer = csv.writer(csvfile)
    
    columns = [("Course#","Professor", "Total_Q", "%Response", "#Contributions", "Avg_Response_Time", "Enrolled", "instructor_response", "%Active", "semester")]
    
    for data in columns:
            writer.writerow(data)
    
    for line in myfile:
        time.sleep(0.01)
        line.rstrip("\n")
        semester = (line[line.find(",")+2:line.rfind(",")])
        url = line[:line.find(",")]
        professor = (line[line.rfind(",")+2:]).rstrip("\n") + " "
        page = requests.get(url)
    
        soup = bs4.BeautifulSoup(page.text, "html.parser")
        
        dataset = []
        
        course = "None"       
        t_q = 0        
        perc_response = 0      
        contributions = 0
        response_time = 0
        date = 0
        enrolled = 0
        instructor_response = 0
        percentage_active = 0
        
        course_class = soup.find("div", {"class": 'title_wrapper'}).h1
        
        course = course_class.text
        
        content_class = soup.find("div", {"class": 'columnContent clearFix'})
        
        q_up_class = content_class.find("div", {"class": 'summaryWithImage'})
        q_up_class_container = q_up_class.find("div", {"class": 'summaryContainer'})
        
        q_class = q_up_class_container.find("div", {"class": 'summaryStatistic'})
        t_q = q_class.span.text
        
        content_list = content_class.select('div[class*="summaryWithImage clearFix"]')       
        
        perc_response_container = content_list[0].find("div", {"class": 'summaryContainer'})      
        perc_container_class_up = perc_response_container.find("div", {"class": 'summaryStatistic'})   
        perc_response = perc_container_class_up.span.text
        
        contribution_container = content_list[1].find("div", {"class": 'summaryContainer'})    
        contribution_container_class_up = contribution_container.find("div", {"class": 'summaryStatistic'}) 
        contributions = contribution_container_class_up.span.text
        
        response_time_container = content_list[2].find("div", {"class": 'summaryContainer'})
        response_time_container_class_up = response_time_container.find("div", {"class": 'summaryStatistic'}) 
        response_time = response_time_container_class_up.span.text
        
        left_side = soup.find("div", {"class": 'column span1'})
        section = left_side.find("div", {"class": 'section'})
        column = section.find("div", {"class": 'columnContent'})
        classParticipation = column.find("div", {"id": 'ClassParticipation'})
        description = classParticipation.select('div[class*="summaryDescription"]')
        percentage_active = description[1].span.text
        enrolled = description[0].span.text
        
        main_side = soup.find("div", {"class": 'column span2'})
        column_content = main_side.find("div", {"class": 'columnContent'})
        graph_section = column_content.find("div", {"class": 'section GraphSection'})
        studentsInstructors = graph_section.find("div", {"class": 'section clearFix studentsInstructors'})
        instructor_response_up_class = studentsInstructors.find("div", {"class": 'summaryWithImage clearFix'})
        instructor_response_container = instructor_response_up_class.find("div", {"class": 'summaryContainer'})
        instructor_response_description = instructor_response_container.find("div", {"class": 'summaryDescription'})
        instructor_response_span_up = instructor_response_description.find("span", {"class": 'inlineStatistic'})      
        instructor_response = instructor_response_span_up.text       
        
        dataset.append((course, professor, t_q, perc_response, contributions, response_time, enrolled, instructor_response, percentage_active, semester))
        
        for data in dataset:
            writer.writerow(data)
    
myfile.close()
```

## [ratemyprofessors.com](https://www.ratemyprofessors.com/)
Information gathered using the raw data in [Rate_my_prof.txt](Rate_my_prof.txt)

```python
myfile = open("Rate_my_prof.txt") 
with open("rmp.csv", "w", newline = '') as csvfile:
    writer = csv.writer(csvfile)
    columns = [("Professor", "Quality", "Difficulty", "Would Take Again", "Date", "Class taught")]
    for data in columns:
            writer.writerow(data)
    for line in myfile:
        url = line
        url = url.rstrip("\n")
        page = requests.get(url)
        
        soup = bs4.BeautifulSoup(page.text, "html.parser")
        
        dataset = []
        
        ul      = soup.find('ul', {'class': 'RatingsList__RatingsUL-hn9one-1 kHITzZ'})
        if ul is not None:
            for item in ul.findAll('li', recursive = False):

                if item is not None:
                    
                    date = "None"
                    quality = "None"
                    difficulty = "None"
                    would_take_again = "None"
                    course = "None"
                
                    name_class = soup.find("div", {"class": 'NameTitle__Name-dowf0z-0 cfjPUG'})
                    first_name = name_class.span.text
                    last_name_class = name_class.find("span", {"class": 'NameTitle__LastNameWrapper-dowf0z-2 glXOHH'})
                    last_name = last_name_class.text
                    name = first_name + " " + last_name
                  
                    q_d_up_class = item.find("div", {"class": 'RatingValues__StyledRatingValues-sc-6dc747-0 bJSTHc'})                    
                    if q_d_up_class is not None:
                        quality_and_difficulty = q_d_up_class.select('div[class*="RatingValues__RatingValue-sc"]')
                        q = str(quality_and_difficulty[0])
                        quality = float(q[-9:-6])
                        d = str(quality_and_difficulty[1])
                        difficulty = float(d[-9:-6])
                    
                    would_take_again_up = item.find("div", {"class": 'CourseMeta__StyledCourseMeta-x344ms-0 fPJDHT'})                    
                    if would_take_again_up is not None:
                        would_take_again_list = would_take_again_up.select('div[class*="MetaItem__StyledMetaItem-"]')
                        if len(would_take_again_list) >2:
                            would_take_again = would_take_again_list[2].span.text
                    time_course_up_class = item.find("div", {"class": 'RatingHeader__StyledHeader-sc-1dlkqw1-0 uBHCj'})
                    if time_course_up_class is not None:
                        time_up_list = time_course_up_class.select('div[class*="TimeStamp__StyledTimeStamp-sc-9q2r30-"]')
                        date = time_up_list[0].text
                        course_up_list = time_course_up_class.select('div[class*="RatingHeader__ClassInfoWrapper-sc-1dlkqw1-1 jxOApy"]')
                        course_up_class = course_up_list[0].find("div",{"class": 'RatingHeader__StyledClass-sc-1dlkqw1-2 gxDIt'})
                        course = course_up_class.text                    

                    dataset.append((name, quality, difficulty, would_take_again, date, course))

            for data in dataset:
                writer.writerow(data)

            time.sleep(0.01)

myfile.close()
```
